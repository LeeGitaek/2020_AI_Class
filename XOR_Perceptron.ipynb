```
#cpu 모드
import torch 
torch.manual_seed(777)
```

```
X = torch.FloatTensor([[0,0],[0,1],[1,0],[1,1]])
Y = torch.FloatTensor([[0],[1],[1],[0]])
```

```
X
```

```
Y
```

NN 모델 정의

```
#모델 정의 
#Layer 정의 
layer1  = torch.nn.Linear(2,2,bias=True)
layer2  = torch.nn.Linear(2,1,bias=True) # fully connected layer 2개입력 / 출력 1
sigmoid = torch.nn.Sigmoid()
```

```
model = torch.nn.Sequential(layer1,sigmoid,layer2,sigmoid) #입력이랑 출력이랑 연결해줌
#순서 중요 
```

```
model
```

모델 학습

```
#Binary Cross Entropy Loss
#매뉴얼 보면서 ㄹ
loss = torch.nn.BCELoss()
```

```
optimizer = torch.optim.SGD(model.parameters(),lr=1)
```

```
for stop in range(50000):
    optimizer.zero_grad()
    hypothesis = model(X)
    cost = loss(hypothesis,Y)
    cost.backward()
    optimizer.step()
    if stop % 100 == 0:
      print(stop,cost.item())
```

모델 평가



```
with torch.no_grad():
    hypothesis = model(X)
    predicted = (hypothesis > 0.5).float()
    accuracy = (predicted == Y).float().mean()
    print('\n Hypothesis : ',hypothesis.numpy(),'\n Correct : ',predicted.numpy(),'\n Accuracy:',accuracy.item())
```
